```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data=np.load("/home/dexter/Documents/GitHub/3D-textures/Experimental/data/variance_res.npy")
```

```python
names=[
    "Bambu P1P pla-",
    "Ender-3 pla-",
    "Ender-3 V3 SE plas-",
    "bambu pla+",
    "Ender-3 pla+",
    "Ender-3 V3 SE pla+",
    "Bambu P1P petg",
    "Ender-3 V3 SE petg",
    "Bambu P1P wood",
    "Formlabs Form 3 (Resin)"
]
new_data=[]
for i in range(len(data)):
    for j in range(len(data)):
        if data[i][j].sum()!=0:
            new_data.append(data[i][j])
            print(names[len(new_data)-1],new_data[-1])

new_data.pop(-2)
names.pop(-2)
new_data=np.array(new_data)

```

```python
new_data = np.array(new_data)

plt.figure(figsize=(8, 5))

plt.boxplot(new_data.T)  # transpose: each feature becomes a box

# feature labels (your "names" list)
if 'names' in globals():
    plt.xticks(
        ticks=np.arange(1, len(names) + 1),
        labels=names,
        rotation=45,
        ha='right'
    )
else:
    plt.xticks(np.arange(1, new_data.shape[1] + 1))

plt.ylabel("Pixel (variance)")
plt.title("Image Variance across each texture")
plt.tight_layout()
plt.savefig("/home/dexter/Documents/GitHub/3D-textures/assets/all.pdf")
plt.show()
```
